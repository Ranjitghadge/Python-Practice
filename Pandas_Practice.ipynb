```python
import pandas as pd

# Create a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'Salary': [50000, 60000, 70000]}
df = pd.DataFrame(data)
```

```python
df
```

```python
df['Salary'] = df['Salary'].apply(lambda x:x+10000)
```

```python
df
```

```python
df['Name'] = df['Name'].apply(lambda x:x.upper())
```

```python
df
```

```python
def net_worth(row):
  salary = row['Salary']
  tax = 0.1*salary
  return salary - tax


df['NetWorth'] = df.apply(net_worth,axis = 1)
```

```python
df
```

```python
df
```

```python
df
```

```python
def cat(Salary):
  if Salary > 60000:
    return 'Civil'
  if Salary > 70000:
    return 'IT'

df['Categories'] = df['Salary'].apply(cat)
```

```python
df
```
